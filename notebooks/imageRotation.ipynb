```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def forceLandscape(image):
    if image.shape[0] > image.shape[1]:
        return np.rot90(image)
    return image
```

```python
img1 = mpimg.imread('sample_images/iss010e07356.jpg')
plt.imshow(img1)
plt.show()
print(img1.shape)
```

```python
rotatedImage1 = forceLandscape(img1)
plt.imshow(rotatedImage1)
plt.show()
print(rotatedImage1.shape)
```
