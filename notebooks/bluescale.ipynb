```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def forceLandscape(image):
    if image.shape[0] > image.shape[1]:
        return np.rot90(image)
    return image
```

```python
img1 = mpimg.imread('sample_images/iss010e07356.jpg')
img1 = forceLandscape(img1)
plt.imshow(img1)
plt.show()
print(img1.shape)
```

```python
img2 = mpimg.imread('sample_images/iss040e102642.jpg')
plt.imshow(img2)
plt.show()
print(img2.shape)
```

```python
def bluescale(image):
    image[:,:,2][image[:,:,0] > image[:,:,2]] = 0
    image[:,:,2][image[:,:,1] > image[:,:,2]] = 0
    return np.dot(image, [0, 0, 1])
```

```python
img1blue = bluescale(img1)
plt.imshow(img1blue, cmap='gray')
plt.show()
print(img1blue.shape)
```

```python
img2blue = bluescale(img2)
plt.imshow(img2blue, cmap='gray')
plt.show()
print(img2blue.shape)
```
