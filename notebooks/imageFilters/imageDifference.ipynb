```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import scipy
from scipy import misc
```

```python
def resizeImage(image, x,y):
    return scipy.misc.imresize(image,(x,y), interp='nearest')
```

```python
img1 = mpimg.imread('../sample_images/_31R2736.jpg')
img1 = resizeImage(img1, 300,500)
plt.imshow(img1)
plt.show()
print(img1.shape)
print('feature size: ' + str(img1.shape[0]*img1.shape[1] * img1.shape[2]))
```

```python
img2 = mpimg.imread('../sample_images/iss052e003678.jpg')
img2 = resizeImage(img2, 300,500)
plt.imshow(img2)
plt.show()
print(img2.shape)
print('feature size: ' + str(img2.shape[0]*img2.shape[1] * img2.shape[2]))
```

```python
img3 = mpimg.imread('../sample_images/iss052e003679.jpg')
img3 = resizeImage(img3, 300,500)
plt.imshow(img3)
plt.show()
print(img3.shape)
print('feature size: ' + str(img3.shape[0]*img3.shape[1] * img3.shape[2]))
```

```python
def imageDifference(image1,image2):
    return image1 - image2
```

```python
i1i2 = imageDifference(img1,img2)
plt.imshow(i1i2)
plt.show()
```

```python
i2i3 = imageDifference(img2,img3)
plt.imshow(i2i3)
plt.show()
```

Messy, maybe grayscale

```python
def grayscale(image):
    return np.dot(image, [0.21, 0.72, 0.07])
```

```python
img1g = grayscale(img1)
img2g = grayscale(img2)
img3g = grayscale(img3)
```

```python
i1i2g = imageDifference(img1g,img2g)
plt.imshow(i1i2g, cmap='gray')
plt.show()
```

```python
i2i3g = imageDifference(img2g,img3g)
plt.imshow(i2i3g, cmap='gray')
plt.show()
```
