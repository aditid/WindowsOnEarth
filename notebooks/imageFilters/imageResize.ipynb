```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import scipy
from scipy import misc
```

```python
img = mpimg.imread('../sample_images/_31R2736.jpg')
plt.imshow(img)
plt.show()
print(img.shape)
print('feature size: ' + str(img.shape[0]*img.shape[1] * img.shape[2]))
```

```python
def resizeImage(image, x,y):
    return scipy.misc.imresize(image,(x,y), interp='nearest')
```

```python
resizeImg = resizeImage(img, 200, 300)
plt.imshow(resizeImg)
plt.show()
print(resizeImg.shape)
print('feature size: ' + str(resizeImg.shape[0]*resizeImg.shape[1] * resizeImg.shape[2]))
```

```python
resizeImg2 = resizeImage(img, 100, 150)
plt.imshow(resizeImg2)
plt.show()
print(resizeImg2.shape)
print('feature size: ' + str(resizeImg2.shape[0]*resizeImg2.shape[1] * resizeImg2.shape[2]))
```
