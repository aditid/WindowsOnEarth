```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('sample_images/_31R2736.jpg')
plt.imshow(img)
plt.show()
print(img.shape)
print('feature size: ' + str(img.shape[0]*img.shape[1] * img.shape[2]))
```

```python
def blackAndWhite(image):
    image = np.dot(image[...,:3], [0.21, 0.72, 0.07])
    image[image < 128] = 0    # Black
    image[image >= 128] = 255 # White
    return image
```

```python
blackAndWhiteImage = blackAndWhite(img)
plt.imshow(blackAndWhiteImage, cmap='gray')
plt.show()
print(blackAndWhiteImage.shape)
print('feature size: ' + str(blackAndWhiteImage.shape[0]*blackAndWhiteImage.shape[1]))
```
