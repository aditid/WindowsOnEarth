```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.21, 0.72, 0.07])
```

```python
img = mpimg.imread('sample_images/_31R2736.jpg')
plt.imshow(img)
plt.show()
```

```python
print(img)
```

```python
gray = rgb2gray(img)
print(gray)
```

```python
plt.imshow(gray)
plt.show()
```
