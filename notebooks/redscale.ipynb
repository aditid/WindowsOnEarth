```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def forceLandscape(image):
    if image.shape[0] > image.shape[1]:
        return np.rot90(image)
    return image
```

```python
img1 = mpimg.imread('sample_images/iss010e07356.jpg')
img1 = forceLandscape(img1)
plt.imshow(img1)
plt.show()
print(img1.shape)
```

```python
img2 = mpimg.imread('sample_images/iss038e029460.jpg')
plt.imshow(img2)
plt.show()
print(img2.shape)
```

```python
def redscale(image):
    image[:,:,0][(image[:,:,0] >= 150) & (image[:,:,1] >= 150) & (image[:,:,2] >= 150)] = 1
    image[:,:,0][image[:,:,1] > image[:,:,0]] = 0
    image[:,:,0][image[:,:,2] > image[:,:,0]] = 0
    return np.dot(image, [1, 0, 0])
```

```python
img1red = redscale(img1)
plt.imshow(img1red, cmap='gray')
plt.show()
print(img1red.shape)
```

```python
img2red = redscale(img2)
plt.imshow(img2red, cmap='gray')
plt.show()
print(img2red.shape)
```
